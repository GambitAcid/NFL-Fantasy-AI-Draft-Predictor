```python
# import dependencies 
import pandas as pd
from bs4 import BeautifulSoup
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import random
import time
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# #website to be scrapped
url = 'https://www.fantasypros.com/nfl/rankings/rookies.php'
browser.visit(url)
```

```python
# find by css
element_list = browser.find_by_css('span[class="fp-icon__icon"]')
```

```python
rookies = []

for each in element_list[0:5]:
    
    each.click()
    time.sleep(random.randint(10,100))
    
    try:
        html = browser.html
        soup = BeautifulSoup(html,'html.parser') 
        table_tags = soup.find_all('div', class_ ="fp-pc__proj-div")
        
        stats = {}
        
        name = soup.find(attrs= {"id": "fp-player-card-title"}).text
        stats['Player'] = name
        
        for tag in table_tags:
            
            label = tag.find('dt', class_ ='fp-pc__label fp-pc__label--proj').text
            value = tag.find('dd',class_="fp-pc__value fp-pc__value--proj").text
            stats[label] = value
         
        rookies.append(stats)
        
    except:  
        pass
    close_button = browser.find_by_css('button[class=" fp-pc__close-button"]')
    close_button[0].click()
    time.sleep(random.randint(10,100))
    
df=pd.DataFrame(rookies)
df
```

```python
# container tag is dl, class " fp-pc__list fp-pc__list--proj"
# table elements: div tag, class=" fp-pc__proj-div"
# label: dt tag, class= " fp-pc__label fp-pc__label--proj"
# value: dd tag, class = " fp-pc__value fp-pc__value--proj"
```

```python
browser.quit()
```

```python
df.to_csv('rookie_projected_stats')
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
import json
import time

res = requests.get("https://www.fantasypros.com/nfl/rankings/rookies.php")
soup = BS(res.content, "html.parser")
```

```python
# var adpData
scripts = soup.find_all("script")
script
```

```python
script = [script for script in scripts if "adpData" in str(script)][0]
script
```

```python
script_lines = str(script).split("\n")
script_lines
```

```python
ecrData = [line for line in script_lines if "ecrData" in line][0]
ecrData
```

```python
player_data = json.loads(ecrData.replace("    var ecrData = ","").replace(";",""))

player_data = player_data['players']

player_data
```

```python
data_list = []

for player in player_data:
    player_name = player['player_name']
    player_id = player['player_id']
    position = player['player_positions']
    team = player['player_team_id']
    age = player['player_age']
    ecr = player['rank_ecr']
    
    pfjsonp = "680"
    api_key = "2efb16f8f36d0897c3e188b67e3292d5"
    url = f"https://partners.fantasypros.com/api/v1/nfl-player.php?api_key={api_key}&callback=pfjsonp_{pfjsonp}&player={player_id}&week=current&scoring=STD"
    res = requests.get(url).text
    player_stats = json.loads(res.replace(f"pfjsonp_{pfjsonp}(","").replace(");",""))
    data = player_stats['stats'][0]['stats']
    data['Player'] = player_name
    data['position'] = position
    data['player_team_id'] = team
    data['player_age'] = age
    data['rank_ecr'] = ecr
    
    data_list.append(data)
```

```python
stats_df = pd.DataFrame(data_list)
stats_df
```

```python
stats_df.to_csv('rookie_stats.csv')
```
